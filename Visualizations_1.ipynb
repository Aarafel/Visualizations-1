```python
#### 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
import plotly.express as px
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x,y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label', rotation='horizontal', ha='right')  # ha = horizontal alignment
plt.title('Line Plot')
plt.show()
```

```python
#### 2. Write a Python program to plot two or more lines on same plot with suitable legends of each line.
x = [1, 2, 3, 4, 5]
y1 =[2, 4, 6, 8, 10]
y2 =[1, 4, 5, 7, 9]
plt.plot(x,y1, label= 'Line1')
plt.plot(x,y2,label= 'Line2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis label', rotation ='horizontal', ha = 'right')
plt.title('Line plot')
plt.legend()
plt.show()
```

```python
#### 3. Write a Python program to plot two or more lines with legends, different widths and colors.
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, label='Line 1', color='blue', linewidth=2)
plt.plot(x,y2, label='Line 2',color='red',linewidth=4)
plt.xlabel('X-axis')
plt.ylabel('Y-label',rotation ='horizontal', ha ='right')
plt.legend()
plt.show()
```

```python
#### 4. Write a Python program to plot two or more lines with different styles.
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, label='Line 1', color='blue', linewidth=2, linestyle='--' )
plt.plot(x,y2, label='Line 2',color='red',linewidth=4,linestyle='-.')
plt.xlabel('X-axis')
plt.ylabel('Y-label',rotation ='horizontal', ha ='right')
plt.legend()
plt.show()
```

```python
#### 5. Write a Python program to plot two or more lines and set the line markers.
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, label='Line 1', color='blue', linewidth=2, linestyle='--',marker='^' )
plt.plot(x,y2, label='Line 2',color='red',linewidth=4,linestyle='-.', marker= 'o')
plt.xlabel('X-axis')
plt.ylabel('Y-label',rotation ='horizontal', ha ='right')
plt.legend()
plt.show()
```

```python
#### 6. Write a Python programming to display a bar chart of the popularity of programming Languages.
    #Sample data:
    #Programming languages: Java, Python, PHP, JavaScript, C#, C++
    #Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

import matplotlib.pyplot as plt

Programmig_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(Programmig_languages, popularity, color='skyblue')

plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

```
